```python
# Create two arrays of 5 arbitrary numbers 
array1 = [5, 2, 7, 8, 2]
array2 = [8, 6, 2, 5, 6]
# Plot two bar charts using plt.bar and the arrays you've created above
# Label them "Example one" and "example two"
plt.bar([1, 3, 5, 7, 9], array1, label="Example one")
plt.bar([2, 4, 6, 8, 10], array2, label="Example two", color='g')
plt.legend()
# Label x axis "Bar number"
plt.xlabel("Bar number")
# Label y axis "Bar height"
plt.ylabel("Bar height")
# Set title "Bar Graph"
plt.title("Bar Graph")
# Show the chart
plt.show()
```

```python
# Histogram Code 
# Create an array of a random normal distribution
# n = height of the bar graph
#bins = way the data is divided into bars
# maxfreq = highest point of the graph - use to set upper graph limit

# clear the plot
plt.clf()

d = np.random.laplace(loc=15, scale=3, size=100)
# Create a histrogram using plt hist()
n, bins, patches = plt.hist(x=d, bins='auto', color='#0504aa',
                            alpha=0.7, rwidth=0.85)
# Set grid for x and y
# set transperancy of the grid using alpha=0.5
plt.grid(axis='y', alpha=0.5)
plt.grid(axis='x', alpha=0.8)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('My First Histogram Ever')
# $ = superscripting -  tells python it is mathematical
plt.text(23, 45, r'$\mu=15, b=3$')
maxfreq = n.max()
# Set a clean upper y-axis limit.
plt.ylim(ymax = maxfreq + 5)
#plt.ylim(ymax=np.ceil(maxfreq / 10) * 10 if maxfreq % 10 else maxfreq + 10)
```
